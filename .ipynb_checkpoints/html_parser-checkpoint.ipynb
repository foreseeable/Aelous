```python
class HtmlHandler(SGMLParser):
    def reset(self):
        self.inTitle = False
        self.inPara = False
        self.result = ""
        self.ret = ""
        SGMLParser.reset(self)

    def start_title(self, attrs):
        self.inTitle = True
        
    def end_title(self):
        self.inTitle = False
        
    def start_p(self, attrs):
        self.inPara = True
        
    def end_p(self):
        self.inPara = False
        
    def handle_data(self, text):
        pflag = False
        if self.inTitle or self.inPara:
            for c in text:
                if c != ' ' and c != '' and c != '\n' and c != '\t':
                    pflag = True
                    break
            if pflag:
                self.result += text
                self.result += " "
            
    def ret_result(self):
        for c in self.result:
            if c != '\n' and c != '\t':
                self.ret += c
        return self.ret
```

```python
if(__name__ == "__main__"):
    input_dir = './html/'
    output_dir = './processed_html/'
    #file_name = 'page7.html'
    Handler = HtmlHandler()
    cnt = 0
    cnt2 = 0
    log_file = open('log2.txt', 'w')
    for file_name in os.listdir(input_dir):
        cnt += 1
        txt_name = file_name[:-4]
        txt_name += 'txt'
        print('processing {}'.format(file_name))
        if file_name == "page8957.html" or file_name == "page4786.html" or file_name == "page4787.html":
            print("pass")
            continue
        if os.path.isfile(os.path.join(output_dir, txt_name)):
            cnt2 += 1
            print('pass')
            continue
        f = open(os.path.join(input_dir, file_name), 'r')
        html = ""
        for tmp in f.readlines():
            tmp.strip('\n')
            html += tmp
        f.close()
        Handler.reset()
        try:
            Handler.feed(html)
        except sgmllib.SGMLParseError:
            log_file.write('failure on {}.\n'.format(file_name))
            print('failure on {}.'.format(file_name))
            continue
        else:
            pass
        finally:
            pass
        result = Handler.ret_result()
        f = open(os.path.join(output_dir, txt_name), 'w')
        f.write(result)
        f.close()
    print("total: {}".format(cnt))
    print("pass cnt: {}".format(cnt2))
    log_file.close()
```
